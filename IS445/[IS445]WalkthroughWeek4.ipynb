```python
# Step 1: hard code a parameter that we want to update interactively
color_map_name = 'hot'

plt.imshow(michigan, cmap=color_map_name)
plt.colorbar()
plt.show()
```

```python
# Step 2: make the parameter an input to a function
#color_map_name = 'hot'

def michigan_plot(color_map_name):
    # COMMAND + ]
    plt.imshow(michigan, cmap=color_map_name)
    plt.colorbar()
    plt.show()
```

```python
# Step 2.5: actually test the function a few times!!
michigan_plot('hot')
```

```python
michigan_plot('cool')
```

```python
# Step 3: decorate the function with an ipywidgets.interact call
# color_map_name is the same as the parameter that is the input to our function!
@ipywidgets.interact(color_map_name=['hot', 'terrain', 'jet']) 
def michigan_plot(color_map_name):
    plt.imshow(michigan, cmap=color_map_name)
    plt.colorbar()
    plt.show()
```
