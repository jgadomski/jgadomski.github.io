```python
#what is the mean, max, min, std of square footage for each year acquired
stats = buildings.groupby('Year Acquired')['Square Footage'].describe()
```

```python
stats
```

```python
type(stats)
```

```python
fig, ax = plt.subplots(figsize=(10,3))
ax.plot(stats['max'], marker='.')
ax.set_xlabel('Year Acquired')
ax.set_ylabel('Max Square Footage')

ax.set_yscale('log')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,3))
ax.plot(stats['max'], marker='.')
ax.plot(stats['min'], marker='.')
ax.set_xlabel('Year Acquired')
ax.set_ylabel('Max and Min Square Footage')

ax.set_yscale('log')
plt.show()
```

```python
with plt.style.context('ggplot'):
    fig, ax = plt.subplots(figsize=(10,3))
    ax.plot(stats['max'], marker='.')
    ax.plot(stats['min'], marker='.')
    ax.set_xlabel('Year Acquired')
    ax.set_ylabel('Max and Min Square Footage')
    ax.set_yscale('log')
    plt.show()
```

```python
def make_plot(style):
    with plt.style.context(style) :
        fig, ax = plt.subplots(figsize=(10,3))
        ax.plot(stats['max'], marker='.')
        ax.plot(stats['min'], marker='.')
        ax.set_xlabel('Year Acquired')
        ax.set_ylabel('Max and Min Square Footage')
        ax.set_yscale('log')
        plt.show()
```

```python
import ipywidgets
```

```python
#decorator - makes it interactive
@ipywidgets.interact(style=plt.style.available)
def make_plot(style):
    with plt.style.context(style) :
        fig, ax = plt.subplots(figsize=(10,3))
        ax.plot(stats['max'], marker='.')
        ax.plot(stats['min'], marker='.')
        ax.set_xlabel('Year Acquired')
        ax.set_ylabel('Max and Min Square Footage')
        ax.set_yscale('log')
        plt.show()
```
