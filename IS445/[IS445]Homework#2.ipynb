```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
import PIL.Image as Image
```

```python
im1 = Image.open("png4.jpeg")
im2 = Image.open('png5.jpeg')
im3 = Image.open('png6.jpeg')
#the reason I wrote png4,5,6 was bc I originally used 3 diff images but chose a whole new set
```

```python
im1
```

```python
im_data1 = np.array(im1)
im_data1.shape 
```

```python
np.unique(im_data1)
```

```python
im_data1.reshape(-1, im_data1.shape[2])
```

```python
np.unique(im_data1.reshape(-1, im_data1.shape[2]),axis=0)
```

```python
plt.imshow(im_data1)
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ax.imshow(im_data1)
plt.show()
```

```python
#fig, ax = plt.subplots(figsize=10,3)
fig, ax = plt.subplots(4, 1, figsize=(20,16))
ax[0].hist(im_data1[:,:,0].flatten(), bins=256, color='red') 
ax[1].hist(im_data1[:,:,1].flatten(), bins=256, color='green') 
ax[2].hist(im_data1[:,:,2].flatten(), bins=256, color='blue') 
ax[3].hist(im_data1[:,:,2].flatten(), bins=256, color='gray') 

#ax.set_xlabel
ax[0].set_xlabel('R channel bins')
ax[1].set_xlabel('G channel bins')
ax[2].set_xlabel('B channel bins')
ax[3].set_xlabel('Alpha channel bins')

#ax.set_ylabel
ax[0].set_ylabel('# of pixels in bin')
ax[1].set_ylabel('# of pixels in bin')
ax[2].set_ylabel('# of pixels in bin')
ax[3].set_ylabel('# of pixels in bin')

plt.show()
```

For the first image I chose, of a yellow bird, the number of pixels in each bin is roughly similar. Since we see a spike of red and green in the 100 levels of the alpha channel, it confirms the accuracy of the pixels in the image. We also see a small spike around level 160 of the G channel and a small spike at level 190 of the R channel. Each histogram appears to be skewed slightly to the right. In order to create yellow using RGB scales, you combine red and green at the same intensity, which is what we are seeing here. This bird has a diverse range of RGB pixels, and the histograms demonstrate that.

For my second image, I chose to stick with the same theme, and go with another bird. This time, a male cardinal. Similar to the last image, we see some large G and R spikes around level 90. Both the R channel and G channel both appear to be skewed right as well. The image confirms what we are seeing in the image, majority red and green, as there is a consistent number of pixels (around 2000 pixels in bin) from levels 100-250 in the R and G channels. The background of green leaves and the bird being red is the reason the histograms have majority R and G pixels. Compared to each other images histograms, the cardinal has clearly the most abundant red pixels, which makes sense.
